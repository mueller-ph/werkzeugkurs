```python
import numpy as np
import pylab as pl

x,y = np.loadtxt('rohr2.dat',unpack=True)

print "Mittelwerte:"

print "x:\t", np.average(x)
print "y:\t", np.average(y), "\n"

print "Standardabweichungen:"

print "x:\t", np.std(x)
print "y:\t", np.std(y), "\n"

print "Correlation: ", np.correlate(x,y)

pl.plot(x,y,'.')
pl.show()
```
