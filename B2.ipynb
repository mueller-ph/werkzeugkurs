```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import numpy as np
data = pd.read_csv('http://www-static.etp.physik.uni-muenchen.de/kurs/Computing/sw/source/D1s.csv')




data['HTore'][data.Heim=="Ein Frankfurt"].plot(kind='hist')
plt.show()
data['ATore'][data.Heim=="Ein Frankfurt"].plot(kind='hist')
plt.show()
```

```python
data['TGes']=(data.HTore+data.ATore)
data['TGes'].plot(kind='hist')
plt.show()
```
