```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import numpy as np

df=pd.read_csv('http://www-static.etp.physik.uni-muenchen.de/kurs/Computing/sw/source/produkt_klima_Tageswerte_19000801_20151231_05792.txt',';\s*',index_col='MESS_DATUM',parse_dates='MESS_DATUM',skip_footer=1,engine='python')

df['LUFTTEMPERATUR']['1945'].plot()
plt.show()
```

```python
dfq=df['LUFTTEMPERATUR'].resample('Q-NOV')
dfq[dfq.index.quarter==3].plot()
plt.show()
```

```python
dfq[(dfq.index.quarter==3) & (dfq.index.year>1970)].plot()
plt.show()
```
